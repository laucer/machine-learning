# Gradient descent:
$L(x) = x^{2} - 4x$

$\frac{d}{dx} = 2x - 4$

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

## Let's plot this function and set strating point to $-5$:



```
x = np.arange(-10, 14, 0.1)
y = x**2-4*x
point = -5

plt.plot(x, y)
plt.plot(point, point**2-4*point, 'ro')
plt.title('Loss function')
```

## It's time to implement gradient

```
def gradient_descent(df= lambda x : 2*x - 4, learning_rate=0.1, current_point = -5, max_iters = 10000, precision=0.000001, debug=False):
  iters = 0
  step_size = 1
  weights = []
  weights.append(current_point)
  while step_size > precision and iters < max_iters:
    previous_point = current_point
    current_point = current_point - learning_rate * df(previous_point)
    step_size = abs(current_point - previous_point)
    iters += 1
    weights.append(current_point)
    if debug:
      print (f'iter: {iters}, step size: {step_size}, point: {current_point}.')
  print(f'Global min: {current_point}.')
  return weights
```

## Let's draw how the gradient descent works

```
plt.figure(figsize=(10,10))
plt.plot(x, y)
weights = gradient_descent()
for weight in weights:
  plt.plot(weight, weight**2 - 4*weight, 'ro')
```
